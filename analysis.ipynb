```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## 2. Load the dataset
df = pd.read_csv('https://data.cms.gov/sites/default/files/2023-04/67157de9-d962-4af0-bf0e-3578b3afec58/inpatient.csv', sep = '|')
```

```python
# Preview of the whole dataset
df
```

```python
# Explore the columns.
for x in df.columns:
    print(x)
```

```python
#  Identify columns related to medical codexes that are ICD codes, DRG codes, or HCPCS codes.
codex_columns = [col for col in df.columns if 'ICD' in col or 'DRG' in col or 'HCPCS' in col]

# Identifying unique codex types
unique_codex_types = set([col.split('_')[0] for col in codex_columns])  # Identify the base codex types

# Display the unique codex types found using a DataFrame
unique_codex_types_df = pd.DataFrame(list(unique_codex_types), columns=['Unique Codex Types'])
display(unique_codex_types_df)
```

```python
## 3. Calculate and show the frequency of unique values in codex-related columns
codex_frequencies = {}
for col in codex_columns:
    codex_frequencies[col] = df[col].value_counts()

for col in codex_columns[:10]:  # Display for the first 10 codex columns
    codex_frequency_df = codex_frequencies[col].reset_index()
    codex_frequency_df.columns = [col, 'Frequency']
    display(codex_frequency_df)  # This will display the frequency counts in a DataFrame
```

```python
# Handling missing values by replacing them with 'Unknown'
for col in codex_columns:
    df[col] = df[col].fillna('Unknown')
```

```python
## 4. Summary of the most common codes found in the dataset (excluding 'Unknown', '0', and empty values)
filtered_codex_frequencies = {}
for col in codex_columns:
    # Convert all values to strings and then strip
    filtered_codex_frequencies[col] = df[
        (df[col].astype(str).str.strip() != 'Unknown') &
        (df[col].notna()) &
        (df[col].astype(str).str.strip() != '0') &
        (df[col].astype(str).str.strip() != '')
    ][col].value_counts()

# Summarize the 20 most common codexes (excluding 'Unknown', '0', empty strings, and null values)
most_common_codexes_filtered = pd.concat([filtered_codex_frequencies[col] for col in codex_columns]).nlargest(20)
most_common_codexes_filtered_df = most_common_codexes_filtered.reset_index()
most_common_codexes_filtered_df.columns = ['Codex', 'Frequency']

# Display the filtered results
display(most_common_codexes_filtered_df)
```

```python
## Patterns and Insights
```

```python
##Top 5 Most Common ICD Diagnosis Codes by State Code

# Group codex data by a region column
codex_by_state = df.groupby('PRVDR_STATE_CD')['ICD_DGNS_CD1'].value_counts()

# Convert the result to a DataFrame for better analysis
codex_by_state_df = codex_by_state.reset_index(name='Count')

# Display the codex data grouped by State Code
display(codex_by_state_df)

# Plotting Top 5 ICD codes by State Code
top_codex_by_state = codex_by_state_df.groupby('PRVDR_STATE_CD').apply(lambda x: x.nlargest(5, 'Count')).reset_index(drop=True)

plt.figure(figsize=(12, 6))
sns.barplot(data=top_codex_by_state, x='PRVDR_STATE_CD', y='Count', hue='ICD_DGNS_CD1')
plt.xticks(rotation=45)
plt.title("Top 5 Most Common ICD Diagnosis Codes by State Code")
plt.show()
```

```python
##Top 5 Most Common ICD Diagnosis Codes Over Time

# Convert the date column to datetime
df['CLM_FROM_DT'] = pd.to_datetime(df['CLM_FROM_DT'])

# Group by year and ICD diagnosis code, and calculate the frequency of each ICD code
codex_by_year = df.groupby(df['CLM_FROM_DT'].dt.year)['ICD_DGNS_CD1'].value_counts()

# Convert to DataFrame for better visualization
codex_by_year_df = codex_by_year.reset_index(name='Count')

# Identify the top 5 most common ICD codes overall
top_5_icd_codes = codex_by_year_df.groupby('ICD_DGNS_CD1')['Count'].sum().nlargest(5).index.tolist()

# Filter the dataset to only include the top 5 ICD codes
top_5_icd_data = codex_by_year_df[codex_by_year_df['ICD_DGNS_CD1'].isin(top_5_icd_codes)]

# Display the codex data grouped by year
display(codex_by_year_df)

# Visualize the top 5 most common ICD diagnosis codes over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=top_5_icd_data, x='CLM_FROM_DT', y='Count', hue='ICD_DGNS_CD1', marker="o")
plt.title("Top 5 Most Common ICD Diagnosis Codes Over Time")
plt.xticks(rotation=45)
plt.show()
```
